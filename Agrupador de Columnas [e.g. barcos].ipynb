```python
import pandas as pd
```

```python
data = pd.read_csv("Plan_Barcos.csv", encoding='latin1', header=[0]) 
```

```python
data.head()
```

```python
keys = [c for c in data if c.startswith('A.')]
data_agrupado = pd.melt(data, id_vars=['IT', 'EMBARCACIÓN'], value_vars=keys, var_name='A_SEMANA', value_name='PORCENTAJE')
data_agrupado.head(10)
```

```python
data_agrupado['SEMANA'] = data_agrupado.A_SEMANA.str.extract(r'\b(\w+)$', expand=True)
```

```python
data_agrupado.drop(['A_SEMANA'], axis='columns', inplace=True)
data_agrupado.head(10)
```

```python
data_agrupado.to_csv("Plan_Barcos-csv.csv", index=False)  # Genera CSV
```
